```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample student lifestyle data (edit values anytime)
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Study_Hours": [3, 4, 2, 5, 3],
    "Mobile_Hours": [5, 4, 6, 3, 4],
    "Sleep_Hours": [6, 7, 5, 8, 6]
}

df = pd.DataFrame(data)

print("Student Daily Data")
display(df)

# Productivity Score
df["Productivity"] = (df["Study_Hours"] * 2 + df["Sleep_Hours"]) - df["Mobile_Hours"]

avg_productivity = df["Productivity"].mean()

print("\nAverage Productivity Score:", round(avg_productivity,2))

# Distraction detection
if avg_productivity < 5:
    status = "Highly Distracted"
elif avg_productivity < 8:
    status = "Moderately Focused"
else:
    status = "Highly Focused"

print("Focus Status:", status)

# Recommendation System
print("\nAI Suggestions:")

if df["Mobile_Hours"].mean() > 4:
    print("ðŸ‘‰ Reduce mobile usage")

if df["Sleep_Hours"].mean() < 7:
    print("ðŸ‘‰ Improve sleep routine")

if df["Study_Hours"].mean() < 4:
    print("ðŸ‘‰ Increase daily study hours")

# Charts
plt.figure()
plt.plot(df["Day"], df["Study_Hours"], label="Study")
plt.plot(df["Day"], df["Mobile_Hours"], label="Mobile")
plt.plot(df["Day"], df["Sleep_Hours"], label="Sleep")
plt.legend()
plt.title("Daily Habit Tracking")
plt.show()

plt.figure()
plt.bar(df["Day"], df["Productivity"])
plt.title("Productivity Score")
plt.show()
```
